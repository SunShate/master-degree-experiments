### Проверка на стационарность

[x] **Augmented Dickey-Fuller (ADF) Test**

It is done with the following assumptions:

- Null Hypothesis (H0): Series is non-stationary
- Alternate Hypothesis (HA): Series is stationary
- p-value > 0.05 Fail to reject (H0)
- p-value <= 0.05 Accept (H1)


[] **Kwiatkowski–Phillips–Schmidt–Shin (KPSS) Test**

The KPSS test is conducted with the following assumptions.

- Null Hypothesis (HO): Series is trend stationary or series has no unit root.
- Alternate Hypothesis(HA): Series is non-stationary, or series has a unit root.
- If the Test Statistic < Critical Value and p-value < 0.05 – Fail to Reject Null Hypothesis(HO), i.e., time series does not have a unit root, meaning it is trend stationary.

*When to Choose ADF or KPSS Test?*

There could be a lot of confusion on when one should use the ADF test or KPSS test and which test would give a correct result. A better solution is to apply/run both tests and makes sure that the series is truly stationary.

The following are the possible outcomes of applying both tests.

- Case 1: Both tests conclude that the given series is stationary – The series is stationary
- Case 2: Both tests conclude that the given series is non-stationary – The series is non-stationary
- Case 3: ADF concludes non-stationary, and KPSS concludes stationary – The series is trend stationary. To make the series strictly stationary, we need to remove the trend in this case. Then we check the detrended series for stationarity.
- Case 4: ADF concludes stationary, and KPSS concludes non-stationary – The series is difference stationary. Differencing is to be used to make series stationary. Then we check the differenced series for stationarity.

```python
def check_stationarity(timeseries):

    # Calculate rolling statistics
    rolling_mean = timeseries.rolling(window=12).mean()
    rolling_std = timeseries.rolling(window=12).std()

    # Plot rolling statistics
    plt.figure(figsize=(10, 6))
    plt.plot(timeseries, color='blue', label='Original')
    plt.plot(rolling_mean, color='red', label='Rolling Mean')
    plt.plot(rolling_std, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()

    # Perform Dickey-Fuller test
    print('Dickey-Fuller Test Results:')
    df_test = adfuller(timeseries)
    df_output = pd.Series(df_test[:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in df_test[4].items():
        df_output[f'Critical Value ({key})'] = value
    print(df_output)
```

Types of Time Series Decomposition Techniques

Additive Decomposition:

- Y(t)=Trend(t) + Seasonal(t) + Residual(t)
- It's suitable when the magnitude of seasonality doesn't vary with the magnitude of the time series.

Multiplicative Decomposition:

- Y(t)=Trend(t) Seasonal(t) Residual(t)
- It's suitable when the magnitude of seasonality scales with the magnitude of the time series.

```python
# Разбиваем данные на 5 частей равного размера и создаем DataFrame
n_splits = 5
split_size = len(data) // n_splits

# Создаем словарь для хранения частей данных
data_parts = {}

# Разбиваем данные и заполняем словарь
for i in range(n_splits):
    start_idx = i * split_size
    end_idx = (i + 1) * split_size if i < n_splits - 1 else len(data)
    data_parts[f'Часть_{i+1}'] = data.iloc[start_idx:end_idx, 1].values

# Создаем DataFrame из словаря с одинаковым количеством строк
max_len = max(len(arr) for arr in data_parts.values())
data_parts_df = pd.DataFrame({
    col: pd.Series(arr, index=range(max_len))
    for col, arr in data_parts.items()
})

```

### K-means

```python
# Импортируем необходимые библиотеки для кластеризации
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

n_clusters = 2
# Подготавливаем данные для кластеризации
X = data_parts_df.values.reshape(-1, 1)

# Создаем pipeline для кластеризации
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('kmeans', KMeans(n_clusters=n_clusters, random_state=RANDOM_SEED))
])

# Обучаем модель и получаем метки кластеров
clusters = pipeline.fit_predict(X)

# Добавляем метки кластеров к исходным данным
data_clustered = data.copy()
data_clustered['Cluster'] = clusters

# Визуализация результатов
plt.figure(figsize=(15, 6))

# График временного ряда с цветовой кодировкой кластеров
plt.subplot(1, 2, 1)
for cluster in range(n_clusters):
    mask = data_clustered['Cluster'] == cluster
    plt.plot(data_clustered[mask].iloc[:, 0], 
             data_clustered[mask].iloc[:, 1], 
             'o-', 
             label=f'Кластер {cluster}',
             markersize=2)
plt.title('Временной ряд с кластерами')
plt.xlabel('Время')
plt.ylabel('Значение')
plt.legend()

# Распределение значений по кластерам (box plot)
plt.subplot(1, 2, 2)
plt.boxplot([data_clustered[data_clustered['Cluster'] == i].iloc[:, 1] 
            for i in range(n_clusters)],
            tick_labels=[f'Кластер {i}' for i in range(n_clusters)])
plt.title('Распределение значений по кластерам')
plt.ylabel('Значение')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from pypsf import Psf
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

class ImputePSF:
    def __init__(self, cycle=12):
        """
        Initialize ImputePSF with cycle length
        Args:
            cycle (int): Length of the seasonal cycle (default: 12 for monthly data)
        """
        self.cycle = cycle

    def _create_missing_profile_matrix(self, data):
        """
        Create missing profile matrix M(X, Y, Z) where:
        X - Start point of missing patch
        Y - End point of missing patch
        Z - Size of missing patch
        """
        if not isinstance(data, (list, np.ndarray, pd.Series)):
            data = data.iloc[:, 0]

        # Find indices of missing values
        missing_indices = np.where(np.isnan(data))[0]
        if len(missing_indices) == 0:
            return pd.DataFrame(columns=['x', 'y', 'z']), 0

        # Find continuous patches
        patches = []
        start_idx = missing_indices[0]
        prev_idx = start_idx

        for i in range(1, len(missing_indices)):
            if missing_indices[i] != prev_idx + 1:
                # End of current patch
                patches.append({
                    'x': start_idx,
                    'y': prev_idx,
                    'z': prev_idx - start_idx + 1
                })
                start_idx = missing_indices[i]
            prev_idx = missing_indices[i]

        # Add last patch
        patches.append({
            'x': start_idx,
            'y': prev_idx,
            'z': prev_idx - start_idx + 1
        })

        # Convert to DataFrame and sort by patch size (z) in descending order
        profile_matrix = pd.DataFrame(patches)
        return profile_matrix.sort_values('z', ascending=False), len(patches)

    def _predict_with_psf(self, data, n_ahead):
        """
        Predict future values using PSF algorithm
        """
        # Remove NaNs and convert to numpy array
        data = np.array(data)
        data = data[~np.isnan(data)]
        
        # If data is too short, use ARIMA directly
        if len(data) < self.cycle * 2:
            try:
                model = ARIMA(data, order=(1, 1, 1))
                fit = model.fit()
                return fit.forecast(n_ahead)
            except Exception as e:
                print(f"ARIMA failed: {e}")
                return np.zeros(n_ahead)

        try:
            # Convert to pandas Series for PSF
            data_series = pd.Series(data)
            psf_model = Psf(cycle_length=self.cycle, 
                           apply_diff=True, 
                           diff_periods=self.cycle)
            psf_model.fit(data_series)
            predictions = psf_model.predict(n_ahead=n_ahead)
            return predictions
        except Exception as e:
            print(f"PSF failed: {e}")
            try:
                # Fallback to ARIMA
                model = ARIMA(data, order=(1, 1, 1))
                fit = model.fit()
                return fit.forecast(n_ahead)
            except Exception as e:
                print(f"ARIMA failed: {e}")
                return np.zeros(n_ahead)

    def impute(self, data):
        """
        Main function to impute missing values in time series
        Args:
            data: Input time series with missing values (numpy array, list, or pandas Series)
        Returns:
            Imputed time series
        """
        # Convert input to numpy array
        D = np.array(data)
        
        # Create missing profile matrix
        M, i = self._create_missing_profile_matrix(D)
        
        # Process each patch in order of size (largest first)
        while i > 0:
            for _, patch in M.iterrows():
                X_j = int(patch['x'])
                Y_j = int(patch['y'])
                Z_j = int(patch['z'])
                print(i)
                
                # Calculate mid point of patch
                mid_point = (X_j + Y_j) / 2
                lens = len(D)
                
                # Determine patch location and imputation strategy
                if mid_point < 0.2 * lens:  # Head section
                    # Use only reverse prediction
                    D_y = D[Y_j:][::-1]  # Reverse data after patch
                    D_Rj = self._predict_with_psf(D_y, Z_j)
                    D_imp = D_Rj[::-1]  # Reverse predictions back
                    
                elif mid_point > 0.8 * lens:  # Tail section
                    # Use only forward prediction
                    D_x = D[:X_j]  # Data before patch
                    D_Fj = self._predict_with_psf(D_x, Z_j)
                    D_imp = D_Fj
                    
                else:  # Body section
                    # Use both forward and reverse predictions
                    D_x = D[:X_j]  # Data before patch
                    D_y = D[Y_j:][::-1]  # Reverse data after patch
                    
                    D_Fj = self._predict_with_psf(D_x, Z_j)
                    D_Rj = self._predict_with_psf(D_y, Z_j)
                    
                    D_imp = (D_Fj + D_Rj[::-1]) / 2
                
                # Replace missing values with imputed values
                n = 0
                for k in range(X_j, Y_j + 1):
                    D[k] = D_imp[n]
                    n += 1
            i -= 1
            
        return D

    def plot_imputation(self, original_data, imputed_data, title="Time Series Imputation"):
        """
        Plot original and imputed data
        """
        plt.figure(figsize=(12, 6))
        
        # Plot original data
        plt.plot(original_data, 'o-', color='red', label='Original Data', alpha=0.5)
        
        # Plot imputed values
        mask = np.isnan(original_data)
        plt.plot(np.where(mask)[0], imputed_data[mask], 'o-', color='blue', label='Imputed Values')
        
        plt.title(title)
        plt.xlabel('Time')
        plt.ylabel('Value')
        plt.legend()
        plt.grid(True)
        plt.show()
```

# Pipeline

```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.impute import KNNImputer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from scipy.interpolate import interp1d
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
from pyVADER.VADER import VADER
from SAITS import SAITS

# Базовый класс для импутеров
class BaseImputer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        raise NotImplementedError

class LinearInterpolationImputer(BaseImputer):
    def transform(self, X):
        x = np.arange(len(X))
        mask = ~np.isnan(X)
        f = interp1d(x[mask], X[mask], kind='linear', fill_value="extrapolate")
        return f(x)

class LOCFImputer(BaseImputer):
    def transform(self, X):
        imputed = np.copy(X)
        last_valid = np.nan
        for i in range(len(imputed)):
            if np.isnan(imputed[i]):
                imputed[i] = last_valid
            else:
                last_valid = imputed[i]
        return imputed

class MovingAverageImputer(BaseImputer):
    def __init__(self, window=3):
        self.window = window
        
    def transform(self, X):
        imputed = pd.Series(X).rolling(window=self.window, center=True).mean().values
        # Заполняем оставшиеся NaN линейной интерполяцией
        return LinearInterpolationImputer().transform(imputed)

class VaDERImputer(BaseImputer):
    def __init__(self, n_clusters=3, latent_dim=2):
        self.n_clusters = n_clusters
        self.latent_dim = latent_dim
        
    def transform(self, X):
        scaler = TimeSeriesScalerMeanVariance()
        scaled_data = scaler.fit_transform(X.reshape(1, -1, 1))
        vader = VADER(n_clusters=self.n_clusters, latent_dim=self.latent_dim)
        imp_vader = vader.fit_transform(scaled_data)
        return scaler.inverse_transform(imp_vader).flatten()

class SAITSImputer(BaseImputer):
    def __init__(self, n_layers=2, d_model=64, d_inner=128, n_head=8):
        self.n_layers = n_layers
        self.d_model = d_model
        self.d_inner = d_inner
        self.n_head = n_head
        
    def transform(self, X):
        saits = SAITS(
            n_layers=self.n_layers,
            d_model=self.d_model,
            d_inner=self.d_inner,
            n_head=self.n_head
        )
        return saits.fit_transform(X.reshape(1, -1, 1)).flatten()

# Создаем словарь пайплайнов для разных методов импутации
pipelines = {
    'Linear Interpolation': Pipeline([
        ('imputer', LinearInterpolationImputer())
    ]),
    'LOCF': Pipeline([
        ('imputer', LOCFImputer())
    ]),
    'Moving Average': Pipeline([
        ('imputer', MovingAverageImputer(window=3))
    ]),
    'KNN': Pipeline([
        ('imputer', KNNImputer(n_neighbors=5))
    ]),
    'VaDER': Pipeline([
        ('imputer', VaDERImputer(n_clusters=3, latent_dim=2))
    ]),
    'SAITS': Pipeline([
        ('imputer', SAITSImputer())
    ])
}

# Функция оценки
def evaluate_imputation(true_data, imputed_data):
    r2 = r2_score(true_data, imputed_data)
    rmse = np.sqrt(mean_squared_error(true_data, imputed_data))
    return r2, rmse

# Подготовка данных
true_data = data.iloc[:, 1].values
missing_data = data_with_missing.iloc[:, 1].values
missing_mask = np.isnan(missing_data)

# Сбор результатов
results = []
for method_name, pipeline in pipelines.items():
    # Применяем импутацию
    imputed_data = pipeline.fit_transform(missing_data.reshape(-1, 1)).ravel()
    
    # Оцениваем только пропущенные значения
    r2, rmse = evaluate_imputation(
        true_data[missing_mask],
        imputed_data[missing_mask]
    )
    
    results.append({
        'method': method_name,
        'r2': r2,
        'rmse': rmse
    })

# Создаем DataFrame с результатами
results_df = pd.DataFrame(results)

# Вывод результатов
print("\nРезультаты импутации:")
for _, row in results_df.iterrows():
    print(f"\n{row['method']}:")
    print(f"  R² = {row['r2']:.4f}")
    print(f"  RMSE = {row['rmse']:.4f}")

# Визуализация результатов
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.bar(results_df['method'], results_df['r2'])
plt.xticks(rotation=45)
plt.ylabel('R²')
plt.title('R² Score по методам импутации')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.bar(results_df['method'], results_df['rmse'])
plt.xticks(rotation=45)
plt.ylabel('RMSE')
plt.title('RMSE по методам импутации')
plt.grid(True)

plt.tight_layout()
plt.show()


```
